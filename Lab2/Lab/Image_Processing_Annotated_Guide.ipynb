## ğŸ”§ åº«å°å…¥èˆ‡åŸºæœ¬æ¦‚å¿µ

### ä»€éº¼æ˜¯åº« (Library)ï¼Ÿ
åº«æ˜¯åˆ¥äººå¯«å¥½çš„ä»£ç¢¼åŒ…ï¼Œæˆ‘å€‘å¯ä»¥ç›´æ¥ä½¿ç”¨å…¶ä¸­çš„å‡½æ•¸ã€‚

```python
# import = å°å…¥åº«çš„é—œéµå­—
# numpy = æ•¸å€¼è¨ˆç®—åº«ï¼Œå¸¸ç”¨ç¸®å¯«æ˜¯ np
# as = çµ¦åº«å–ä¸€å€‹ç°¡çŸ­çš„åˆ¥å
import numpy as np

# matplotlib.pyplot = ç¹ªåœ–åº«ï¼Œå¸¸ç”¨ç¸®å¯«æ˜¯ plt
import matplotlib.pyplot as plt

# from...import = å¾æŸå€‹åº«ä¸­åªå°å…¥ç‰¹å®šéƒ¨åˆ†
# skimage = å°ˆé–€è™•ç†åœ–åƒçš„åº«
# data = skimageä¸­åŒ…å«ç¯„ä¾‹åœ–åƒçš„æ¨¡çµ„
from skimage import data

# try...except = å˜—è©¦åŸ·è¡Œä»£ç¢¼ï¼Œå¦‚æœå¤±æ•—å°±åŸ·è¡Œåˆ¥çš„
try:
    import cv2  # OpenCVåœ–åƒè™•ç†åº«
    print("âœ… OpenCV å®‰è£æˆåŠŸï¼")
    print(f"ç‰ˆæœ¬: {cv2.__version__}")  # __version__ = é¡¯ç¤ºç‰ˆæœ¬è™Ÿ
except ImportError:  # ImportError = å°å…¥å¤±æ•—çš„éŒ¯èª¤é¡å‹
    print("âš ï¸ OpenCV æœªå®‰è£")

# print() = é¡¯ç¤ºæ–‡å­—åˆ°è¢å¹•çš„å‡½æ•¸
print("æ‰€æœ‰åº«å°å…¥å®Œæˆï¼")
```

## 1ï¸âƒ£ å‰µå»ºç¬¬ä¸€å€‹"åœ–åƒ" - ç†è§£æ•¸çµ„

### ğŸ¯ é—œéµæ¦‚å¿µè§£é‡‹
- **np.array()** = numpyçš„å‡½æ•¸ï¼Œç”¨ä¾†å‰µå»ºæ•¸çµ„(é™£åˆ—)
- **dtype** = data typeçš„ç¸®å¯«ï¼ŒæŒ‡å®šæ•¸æ“šé¡å‹
- **uint8** = 8ä½ç„¡ç¬¦è™Ÿæ•´æ•¸ï¼Œç¯„åœ0-255ï¼Œæ˜¯åœ–åƒå¸¸ç”¨çš„æ•¸æ“šæ ¼å¼

```python
# np.array() = å‰µå»ºnumpyæ•¸çµ„çš„å‡½æ•¸
# [ ] = æ–¹æ‹¬è™Ÿè¡¨ç¤ºåˆ—è¡¨
# [[ ]] = äºŒç¶­åˆ—è¡¨ï¼Œè¡¨ç¤ºçŸ©é™£çš„è¡Œ
simple_image = np.array([
    [0,   128, 255],  # ç¬¬ä¸€è¡Œï¼šé»‘è‰², ç°è‰², ç™½è‰²
    [64,  192, 128],  # ç¬¬äºŒè¡Œï¼šæ·±ç°, æ·ºç°, ä¸­ç°
    [255, 0,   64]    # ç¬¬ä¸‰è¡Œï¼šç™½è‰², é»‘è‰², æ·±ç°
], dtype=np.uint8)    # dtype = æŒ‡å®šæ•¸æ“šé¡å‹ç‚º uint8

# print() = é¡¯ç¤ºå…§å®¹çš„å‡½æ•¸
print("æˆ‘å€‘å‰µå»ºçš„3x3åœ–åƒçŸ©é™£:")
print(simple_image)  # é¡¯ç¤ºæ•¸çµ„å…§å®¹

# .shape = æ•¸çµ„çš„å±¬æ€§ï¼Œé¡¯ç¤ºç¶­åº¦å¤§å°
print(f"å½¢ç‹€: {simple_image.shape}")  # f"..." = æ ¼å¼åŒ–å­—ç¬¦ä¸²

# .dtype = æ•¸çµ„çš„å±¬æ€§ï¼Œé¡¯ç¤ºæ•¸æ“šé¡å‹
print(f"æ•¸æ“šé¡å‹: {simple_image.dtype}")

# .min() = æ•¸çµ„çš„æ–¹æ³•ï¼Œæ‰¾æœ€å°å€¼
# .max() = æ•¸çµ„çš„æ–¹æ³•ï¼Œæ‰¾æœ€å¤§å€¼
print(f"åƒç´ å€¼ç¯„åœ: {simple_image.min()} åˆ° {simple_image.max()}")
```

```python
# === matplotlib ç¹ªåœ–è©³ç´°è§£é‡‹ ===

# plt.figure() = å‰µå»ºä¸€å€‹æ–°çš„åœ–å½¢çª—å£
# figsize = åœ–å½¢å¤§å°ï¼Œæ ¼å¼æ˜¯ (å¯¬åº¦, é«˜åº¦)ï¼Œå–®ä½æ˜¯è‹±å¯¸
plt.figure(figsize=(8, 4))

# plt.subplot() = åœ¨åœ–å½¢ä¸­å‰µå»ºå­åœ–
# æ ¼å¼: (è¡Œæ•¸, åˆ—æ•¸, ç•¶å‰ä½ç½®)
# (1, 2, 1) = 1è¡Œ2åˆ—ï¼Œé¸ç¬¬1å€‹ä½ç½®
plt.subplot(1, 2, 1)

# plt.imshow() = é¡¯ç¤ºåœ–åƒçš„å‡½æ•¸
# simple_image = è¦é¡¯ç¤ºçš„åœ–åƒæ•¸æ“š
# cmap = colormapçš„ç¸®å¯«ï¼Œé¡è‰²æ˜ å°„æ–¹æ¡ˆ
# 'gray' = ç°åº¦é¡è‰²æ–¹æ¡ˆï¼Œé»‘ç™½é¡¯ç¤º
# vmin, vmax = é¡è‰²ç¯„åœçš„æœ€å°å€¼å’Œæœ€å¤§å€¼
plt.imshow(simple_image, cmap='gray', vmin=0, vmax=255)

# plt.title() = è¨­ç½®åœ–è¡¨æ¨™é¡Œ
plt.title('æˆ‘çš„ç¬¬ä¸€å€‹3x3åœ–åƒ')

# plt.colorbar() = æ·»åŠ é¡è‰²æ¢ï¼Œé¡¯ç¤ºæ•¸å€¼å°æ‡‰çš„é¡è‰²
# label = é¡è‰²æ¢çš„æ¨™ç±¤æ–‡å­—
plt.colorbar(label='åƒç´ å€¼ (0=é»‘è‰², 255=ç™½è‰²)')

# å‰µå»ºç¬¬äºŒå€‹å­åœ–
plt.subplot(1, 2, 2)  # 1è¡Œ2åˆ—ï¼Œé¸ç¬¬2å€‹ä½ç½®

# interpolation = æ’å€¼æ–¹æ³•ï¼Œæ±ºå®šåƒç´ ä¹‹é–“å¦‚ä½•å¡«å……
# 'nearest' = æœ€è¿‘é„°æ’å€¼ï¼Œä¿æŒåƒç´ çš„æ–¹å¡Šç‹€
plt.imshow(simple_image, cmap='gray', vmin=0, vmax=255, interpolation='nearest')
plt.title('æ”¾å¤§é¡¯ç¤º (ä¿æŒæ–¹å¡Šç‹€)')

# æ·»åŠ ç¶²æ ¼ç·šå¹«åŠ©ç†è§£åƒç´ ä½ç½®
# range(4) = ç”Ÿæˆ 0, 1, 2, 3 çš„æ•¸åˆ—
for i in range(4):  # for = å¾ªç’°ï¼Œé‡è¤‡åŸ·è¡Œ
    # plt.axhline() = ç•«æ°´å¹³ç·š
    # i-0.5 = ç·šçš„yä½ç½®
    # color = ç·šçš„é¡è‰²
    # linewidth = ç·šçš„ç²—ç´°
    plt.axhline(i-0.5, color='red', linewidth=1)
    # plt.axvline() = ç•«å‚ç›´ç·š
    plt.axvline(i-0.5, color='red', linewidth=1)

# plt.tight_layout() = è‡ªå‹•èª¿æ•´å­åœ–é–“è·ï¼Œé¿å…é‡ç–Š
plt.tight_layout()

# plt.show() = é¡¯ç¤ºåœ–å½¢
plt.show()

print("ğŸ” é‡è¦æ¦‚å¿µ: æ•¸å­—è¶Šå¤§ï¼Œåƒç´ è¶Šäº®")
print("0 = ç´”é»‘è‰²")
print("128 = ä¸­ç°è‰²")
print("255 = ç´”ç™½è‰²")
```

## 2ï¸âƒ£ è¼‰å…¥çœŸå¯¦åœ–åƒ

### ğŸ¯ é—œéµæ¦‚å¿µè§£é‡‹
- **data.astronaut()** = skimageåº«å…§å»ºçš„å¤ªç©ºäººåœ–åƒ
- **shapeå±¬æ€§** = é¡¯ç¤ºåœ–åƒå°ºå¯¸ï¼š(é«˜åº¦, å¯¬åº¦, é€šé“æ•¸)
- **RGBåœ–åƒ** = æœ‰3å€‹é¡è‰²é€šé“ï¼šRed(ç´…), Green(ç¶ ), Blue(è—)

```python
# data.astronaut() = è¼‰å…¥å…§å»ºçš„å¤ªç©ºäººåœ–åƒ
# é€™æ˜¯ä¸€å€‹ç¶“å…¸çš„æ¸¬è©¦åœ–åƒï¼Œå¤§å°æ˜¯ 512x512x3
astronaut_rgb = data.astronaut()

print("=== åœ–åƒåŸºæœ¬è³‡è¨Šåˆ†æ ===")

# .shape = åœ–åƒçš„ç¶­åº¦è³‡è¨Š
# å°æ–¼å½©è‰²åœ–åƒï¼Œæ ¼å¼æ˜¯ (é«˜åº¦, å¯¬åº¦, é€šé“æ•¸)
print(f"åœ–åƒå½¢ç‹€: {astronaut_rgb.shape}")
print(f"  - é«˜åº¦: {astronaut_rgb.shape[0]} åƒç´ ")
print(f"  - å¯¬åº¦: {astronaut_rgb.shape[1]} åƒç´ ")
print(f"  - é€šé“æ•¸: {astronaut_rgb.shape[2]} (RGBä¸‰é€šé“)")

# .dtype = æ•¸æ“šé¡å‹
print(f"æ•¸æ“šé¡å‹: {astronaut_rgb.dtype}")

# .min() å’Œ .max() = æ‰€æœ‰åƒç´ ä¸­çš„æœ€å°å€¼å’Œæœ€å¤§å€¼
print(f"åƒç´ å€¼ç¯„åœ: {astronaut_rgb.min()} åˆ° {astronaut_rgb.max()}")

# .size = ç¸½åƒç´ æ•¸ (é•·xå¯¬xé€šé“)
print(f"ç¸½åƒç´ æ•¸: {astronaut_rgb.size}")

# .nbytes = è¨˜æ†¶é«”ä½¿ç”¨é‡ï¼Œå–®ä½æ˜¯ bytesï¼ˆå­—ç¯€ï¼‰
print(f"è¨˜æ†¶é«”ä½¿ç”¨: {astronaut_rgb.nbytes} bytes")
print(f"  ç›¸ç•¶æ–¼: {astronaut_rgb.nbytes / 1024:.1f} KB")
```

```python
# é¡¯ç¤ºè¼‰å…¥çš„åœ–åƒ

# å‰µå»ºåœ–å½¢ï¼Œè¨­å®šå¤§å°ç‚º8x6è‹±å¯¸
plt.figure(figsize=(8, 6))

# é¡¯ç¤ºRGBå½©è‰²åœ–åƒ
# æ³¨æ„ï¼šå°æ–¼RGBåœ–åƒï¼Œä¸éœ€è¦æŒ‡å®šcmap
plt.imshow(astronaut_rgb)

# è¨­ç½®æ¨™é¡Œ
plt.title('å¤ªç©ºäººåœ–åƒ - RGBå½©è‰² (512x512x3)')

# plt.axis('off') = éšè—åæ¨™è»¸å’Œåˆ»åº¦
# é€™æ¨£åœ–åƒçœ‹èµ·ä¾†æ›´ä¹¾æ·¨
plt.axis('off')

# é¡¯ç¤ºåœ–å½¢
plt.show()

print("ğŸš€ é€™æ˜¯ä¸€å€‹RGBå½©è‰²åœ–åƒï¼")
print("ğŸ“ RGB = Red(ç´…) + Green(ç¶ ) + Blue(è—)")
```

## 3ï¸âƒ£ æ·±å…¥ç†è§£RGBé¡è‰²ç³»çµ±

### ğŸ¯ é—œéµæ¦‚å¿µè§£é‡‹
- **åˆ‡ç‰‡èªæ³• [è¡Œ, åˆ—, é€šé“]** = å¾æ•¸çµ„ä¸­æå–ç‰¹å®šéƒ¨åˆ†
- **[:, :, 0]** = æ‰€æœ‰è¡Œï¼Œæ‰€æœ‰åˆ—ï¼Œç¬¬0å€‹é€šé“ï¼ˆç´…è‰²ï¼‰
- **[:, :, 1]** = æ‰€æœ‰è¡Œï¼Œæ‰€æœ‰åˆ—ï¼Œç¬¬1å€‹é€šé“ï¼ˆç¶ è‰²ï¼‰
- **[:, :, 2]** = æ‰€æœ‰è¡Œï¼Œæ‰€æœ‰åˆ—ï¼Œç¬¬2å€‹é€šé“ï¼ˆè—è‰²ï¼‰

```python
# === æ•¸çµ„åˆ‡ç‰‡èªæ³•è©³ç´°è§£é‡‹ ===

print("=== å­¸ç¿’æ•¸çµ„åˆ‡ç‰‡ ===")
print("åˆ‡ç‰‡èªæ³•: array[è¡Œ, åˆ—, é€šé“]")
print(": = å†’è™Ÿè¡¨ç¤º'æ‰€æœ‰'")
print("[:, :, 0] = æ‰€æœ‰è¡Œ, æ‰€æœ‰åˆ—, ç¬¬0é€šé“")

# åˆ†é›¢RGBä¸‰å€‹é€šé“
# [:, :, 0] = æ‰€æœ‰è¡Œï¼Œæ‰€æœ‰åˆ—ï¼Œç¬¬0å€‹é€šé“ï¼ˆç´…è‰²ï¼‰
red_channel = astronaut_rgb[:, :, 0]

# [:, :, 1] = æ‰€æœ‰è¡Œï¼Œæ‰€æœ‰åˆ—ï¼Œç¬¬1å€‹é€šé“ï¼ˆç¶ è‰²ï¼‰
green_channel = astronaut_rgb[:, :, 1]

# [:, :, 2] = æ‰€æœ‰è¡Œï¼Œæ‰€æœ‰åˆ—ï¼Œç¬¬2å€‹é€šé“ï¼ˆè—è‰²ï¼‰
blue_channel = astronaut_rgb[:, :, 2]

print("\n=== RGBé€šé“è³‡è¨Š ===")
print(f"åŸå§‹å½©è‰²åœ–åƒå½¢ç‹€: {astronaut_rgb.shape}")
print(f"ç´…è‰²é€šé“å½¢ç‹€: {red_channel.shape}")  # åªæœ‰2å€‹ç¶­åº¦äº†
print(f"ç¶ è‰²é€šé“å½¢ç‹€: {green_channel.shape}")
print(f"è—è‰²é€šé“å½¢ç‹€: {blue_channel.shape}")
print("\nğŸ’¡ æ³¨æ„: å–®å€‹é€šé“æ˜¯ç°åº¦åœ–åƒï¼Œæ²’æœ‰é¡è‰²ç¶­åº¦")
```

```python
# é¡¯ç¤ºRGBä¸‰å€‹é€šé“

# å‰µå»º1è¡Œ4åˆ—çš„å­åœ–å¸ƒå±€
plt.figure(figsize=(16, 4))  # å¯¬ä¸€äº›ï¼Œå› ç‚ºæœ‰4å€‹åœ–

# ç¬¬1å€‹å­åœ–ï¼šåŸå§‹RGBåœ–åƒ
plt.subplot(1, 4, 1)  # 1è¡Œ4åˆ—ï¼Œç¬¬1å€‹ä½ç½®
plt.imshow(astronaut_rgb)  # RGBåœ–åƒç›´æ¥é¡¯ç¤º
plt.title('åŸå§‹ RGB åœ–åƒ')
plt.axis('off')  # ä¸é¡¯ç¤ºåæ¨™è»¸

# ç¬¬2å€‹å­åœ–ï¼šç´…è‰²é€šé“
plt.subplot(1, 4, 2)  # 1è¡Œ4åˆ—ï¼Œç¬¬2å€‹ä½ç½®
# cmap='Reds' = ä½¿ç”¨ç´…è‰²çš„é¡è‰²æ˜ å°„
# é€™æ¨£æ•¸å€¼å¤§çš„åœ°æ–¹é¡¯ç¤ºæ›´ç´…
plt.imshow(red_channel, cmap='Reds')
plt.title('ç´…è‰²é€šé“\n(Rå€¼è¶Šå¤§è¶Šäº®)')
plt.axis('off')

# ç¬¬3å€‹å­åœ–ï¼šç¶ è‰²é€šé“
plt.subplot(1, 4, 3)  # 1è¡Œ4åˆ—ï¼Œç¬¬3å€‹ä½ç½®
# cmap='Greens' = ä½¿ç”¨ç¶ è‰²çš„é¡è‰²æ˜ å°„
plt.imshow(green_channel, cmap='Greens')
plt.title('ç¶ è‰²é€šé“\n(Gå€¼è¶Šå¤§è¶Šäº®)')
plt.axis('off')

# ç¬¬4å€‹å­åœ–ï¼šè—è‰²é€šé“
plt.subplot(1, 4, 4)  # 1è¡Œ4åˆ—ï¼Œç¬¬4å€‹ä½ç½®
# cmap='Blues' = ä½¿ç”¨è—è‰²çš„é¡è‰²æ˜ å°„
plt.imshow(blue_channel, cmap='Blues')
plt.title('è—è‰²é€šé“\n(Bå€¼è¶Šå¤§è¶Šäº®)')
plt.axis('off')

# è‡ªå‹•èª¿æ•´å¸ƒå±€
plt.tight_layout()
plt.show()

print("ğŸ¨ é‡è¦è§€å¯Ÿ:")
print("- æ¯å€‹é€šé“éƒ½æ˜¯ä¸€å€‹ç°åº¦åœ–åƒ")
print("- äº®çš„åœ°æ–¹è¡¨ç¤ºè©²é¡è‰²æˆåˆ†å¤š")
print("- æš—çš„åœ°æ–¹è¡¨ç¤ºè©²é¡è‰²æˆåˆ†å°‘")
print("- ä¸‰å€‹é€šé“çµ„åˆèµ·ä¾†å°±æ˜¯å½©è‰²åœ–åƒ")
```

### ğŸ’¡ å¯¦é©—ï¼šå‰µå»ºç´”è‰²åœ–åƒä¾†ç†è§£RGB

```python
# === å‰µå»ºç´”è‰²åœ–åƒå¯¦é©— ===

print("=== å‰µå»ºç´”è‰²åœ–åƒä¾†ç†è§£RGB ===")

# å®šç¾©åœ–åƒå¤§å°
height, width = 100, 100  # 100x100 åƒç´ 

# np.zeros() = å‰µå»ºå…¨é›¶çš„æ•¸çµ„
# (height, width, 3) = 100x100x3 çš„ä¸‰ç¶­æ•¸çµ„
# dtype=np.uint8 = ä½¿ç”¨8ä½æ•´æ•¸æ ¼å¼

# å‰µå»ºç´”ç´…è‰²åœ–åƒ
pure_red = np.zeros((height, width, 3), dtype=np.uint8)
pure_red[:, :, 0] = 255  # ç´…è‰²é€šé“è¨­ç‚º255ï¼ˆæœ€å¤§å€¼ï¼‰
# ç¶ è‰²å’Œè—è‰²é€šé“ä¿æŒ0ï¼ˆæœ€å°å€¼ï¼‰

# å‰µå»ºç´”ç¶ è‰²åœ–åƒ
pure_green = np.zeros((height, width, 3), dtype=np.uint8)
pure_green[:, :, 1] = 255  # åªè¨­ç¶ è‰²é€šé“ç‚º255

# å‰µå»ºç´”è—è‰²åœ–åƒ
pure_blue = np.zeros((height, width, 3), dtype=np.uint8)
pure_blue[:, :, 2] = 255  # åªè¨­è—è‰²é€šé“ç‚º255

# å‰µå»ºæ··åˆé¡è‰²ï¼šé»ƒè‰² = ç´…è‰² + ç¶ è‰²
yellow = np.zeros((height, width, 3), dtype=np.uint8)
yellow[:, :, 0] = 255  # ç´…è‰²é€šé“ = 255
yellow[:, :, 1] = 255  # ç¶ è‰²é€šé“ = 255
# è—è‰²é€šé“ = 0ï¼ˆä¿æŒé»˜èªï¼‰

print("é¡è‰²é…æ–¹:")
print("ç´”ç´… = (255, 0, 0)")
print("ç´”ç¶  = (0, 255, 0)")
print("ç´”è— = (0, 0, 255)")
print("é»ƒè‰² = (255, 255, 0) = ç´… + ç¶ ")
```

```python
# é¡¯ç¤ºå‰µå»ºçš„ç´”è‰²åœ–åƒ

plt.figure(figsize=(16, 4))  # å¯¬åœ–ï¼Œé¡¯ç¤º4å€‹é¡è‰²

# ä½¿ç”¨åˆ—è¡¨ä¾†ç°¡åŒ–ä»£ç¢¼
# zip() = å°‡å¤šå€‹åˆ—è¡¨é…å°
colors = [pure_red, pure_green, pure_blue, yellow]  # åœ–åƒåˆ—è¡¨
titles = [  # æ¨™é¡Œåˆ—è¡¨
    'ç´”ç´…è‰²\nRGB(255,0,0)', 
    'ç´”ç¶ è‰²\nRGB(0,255,0)', 
    'ç´”è—è‰²\nRGB(0,0,255)', 
    'é»ƒè‰²\nRGB(255,255,0)'
]

# enumerate() = åŒæ™‚ç²å¾—ç´¢å¼•å’Œæ•¸å€¼
# zip() = å°‡colorså’Œtitlesé…å°
for i, (color, title) in enumerate(zip(colors, titles)):
    plt.subplot(1, 4, i+1)  # i+1 å› ç‚ºsubplotå¾1é–‹å§‹è¨ˆç®—
    plt.imshow(color)       # é¡¯ç¤ºé¡è‰²åœ–åƒ
    plt.title(title)        # è¨­ç½®æ¨™é¡Œ
    plt.axis('off')         # éš±è—åæ¨™è»¸

plt.tight_layout()
plt.show()

print("ğŸŒˆ RGBé¡è‰²æ··åˆè¦å‰‡:")
print("- RGBæ˜¯åŠ æ³•æ··è‰²ç³»çµ±")
print("- ç´… + ç¶  = é»ƒ")
print("- ç´… + è— = æ´‹ç´…(å“ç´…)")
print("- ç¶  + è— = é’")
print("- ç´… + ç¶  + è— = ç™½")
```

## 4ï¸âƒ£ å½©è‰²åœ–åƒè½‰ç°åº¦åœ–åƒ

### ğŸ¯ é—œéµæ¦‚å¿µè§£é‡‹
- **ç°åº¦è½‰æ›** = å°‡å½©è‰²åœ–åƒè®Šæˆé»‘ç™½åœ–åƒ
- **æ¬Šé‡å…¬å¼** = Gray = 0.299Ã—R + 0.587Ã—G + 0.114Ã—B
- **np.dot()** = æ•¸å­¸ä¸­çš„é»ç©é‹ç®—ï¼Œç”¨ä¾†é«˜æ•ˆè¨ˆç®—æ¬Šé‡å’Œ
- **äººçœ¼æ•æ„Ÿåº¦** = äººçœ¼å°ç¶ å…‰æœ€æ•æ„Ÿï¼Œè—å…‰æœ€ä¸æ•æ„Ÿ

```python
# === ç°åº¦è½‰æ›çš„æ¨™æº–å…¬å¼ ===

print("=== ç°åº¦è½‰æ›æ¬Šé‡è§£é‡‹ ===")

# æ¨™æº–çš„ç°åº¦è½‰æ›æ¬Šé‡ï¼ˆåŸºæ–¼äººçœ¼æ•æ„Ÿåº¦ï¼‰
weights = [0.299, 0.587, 0.114]  # [ç´…è‰²æ¬Šé‡, ç¶ è‰²æ¬Šé‡, è—è‰²æ¬Šé‡]

print("æ¨™æº–ç°åº¦è½‰æ›å…¬å¼:")
print(f"Gray = {weights[0]}Ã—R + {weights[1]}Ã—G + {weights[2]}Ã—B")
print()

# ç‚ºä»€éº¼è¦ç”¨é€™äº›æ¬Šé‡ï¼Ÿ
print("æ¬Šé‡åˆ†é…çš„ç§‘å­¸åŸç†:")
print(f"ç´…è‰²æ¬Šé‡: {weights[0]} ({weights[0]*100:.1f}%) - ä¸­ç­‰æ•æ„Ÿåº¦")
print(f"ç¶ è‰²æ¬Šé‡: {weights[1]} ({weights[1]*100:.1f}%) - æœ€é«˜æ•æ„Ÿåº¦")
print(f"è—è‰²æ¬Šé‡: {weights[2]} ({weights[2]*100:.1f}%) - æœ€ä½æ•æ„Ÿåº¦")
print(f"ç¸½å’Œæª¢æŸ¥: {sum(weights)} (å¿…é ˆç­‰æ–¼1)")
print()

print("ğŸ’¡ ç‚ºä»€éº¼é€™æ¨£åˆ†é…ï¼Ÿ")
print("- äººçœ¼å°ç¶ å…‰æœ€æ•æ„Ÿï¼ˆç™½å¤©è¦–è¦ºï¼‰")
print("- å°ç´…å…‰ä¸­ç­‰æ•æ„Ÿ")
print("- å°è—å…‰æœ€ä¸æ•æ„Ÿ")
print("- é€™æ¨£è½‰æ›çš„ç°åº¦åœ–æœ€æ¥è¿‘äººçœ¼æ„ŸçŸ¥")
```

```python
# === è©³ç´°åˆ†è§£ np.dot() é‹ç®— ===

print("=== ç†è§£ np.dot() é»ç©é‹ç®— ===")

print(f"åŸå§‹RGBåœ–åƒå½¢ç‹€: {astronaut_rgb.shape}")  # (512, 512, 3)
print(f"æ¬Šé‡æ•¸çµ„å½¢ç‹€: {np.array(weights).shape}")   # (3,)
print()

# æ–¹æ³•1: æ‰‹å‹•è¨ˆç®—ï¼ˆæ­¥é©Ÿæ¸…æ™°ä½†æ…¢ï¼‰
print("æ–¹æ³•1: æ‰‹å‹•é€æ­¥è¨ˆç®—")
print("æ­¥é©Ÿåˆ†è§£:")
red_part = astronaut_rgb[:, :, 0] * weights[0]    # ç´…è‰²éƒ¨åˆ†
green_part = astronaut_rgb[:, :, 1] * weights[1]  # ç¶ è‰²éƒ¨åˆ†
blue_part = astronaut_rgb[:, :, 2] * weights[2]   # è—è‰²éƒ¨åˆ†

print(f"1. ç´…è‰² Ã— {weights[0]} = å½¢ç‹€ {red_part.shape}")
print(f"2. ç¶ è‰² Ã— {weights[1]} = å½¢ç‹€ {green_part.shape}")
print(f"3. è—è‰² Ã— {weights[2]} = å½¢ç‹€ {blue_part.shape}")

# ä¸‰éƒ¨åˆ†ç›¸åŠ 
gray_manual = (red_part + green_part + blue_part).astype(np.uint8)
print(f"4. ä¸‰éƒ¨åˆ†ç›¸åŠ  = å½¢ç‹€ {gray_manual.shape}")
print()

# æ–¹æ³•2: ä½¿ç”¨ np.dot()ï¼ˆå¿«é€Ÿï¼‰
print("æ–¹æ³•2: ä½¿ç”¨ np.dot() å¿«é€Ÿè¨ˆç®—")
# [...,:3] = å–æ‰€æœ‰è¡Œã€æ‰€æœ‰åˆ—ã€å‰3å€‹é€šé“ï¼ˆé˜²æ­¢æœ‰ç¬¬4å€‹é€æ˜åº¦é€šé“ï¼‰
gray_dot = np.dot(astronaut_rgb[...,:3], weights).astype(np.uint8)
print(f"np.dot() çµæœå½¢ç‹€: {gray_dot.shape}")
print()

# é©—è­‰å…©ç¨®æ–¹æ³•çµæœç›¸åŒ
# np.array_equal() = æª¢æŸ¥å…©å€‹æ•¸çµ„æ˜¯å¦å®Œå…¨ç›¸åŒ
print(f"å…©ç¨®æ–¹æ³•çµæœç›¸åŒ: {np.array_equal(gray_manual, gray_dot)}")
print("âœ… np.dot() å°±æ˜¯è‡ªå‹•åŒ–çš„åŠ æ¬Šæ±‚å’Œï¼")
```

```python
# === æ¯”è¼ƒä¸åŒçš„ç°åº¦è½‰æ›æ–¹æ³• ===

print("=== æ¯”è¼ƒä¸åŒç°åº¦è½‰æ›æ–¹æ³• ===")

# æ–¹æ³•1: æ¨™æº–åŠ æ¬Šå¹³å‡ï¼ˆæ¨è–¦ï¼‰
gray_standard = np.dot(astronaut_rgb[...,:3], [0.299, 0.587, 0.114]).astype(np.uint8)

# æ–¹æ³•2: ç°¡å–®å¹³å‡
# np.mean() = æ±‚å¹³å‡å€¼
# axis=2 = æ²¿è‘—ç¬¬2å€‹è»¸ï¼ˆé€šé“è»¸ï¼‰æ±‚å¹³å‡
gray_average = np.mean(astronaut_rgb, axis=2).astype(np.uint8)

# æ–¹æ³•3: å–æœ€å¤§å€¼
# np.max() = æ±‚æœ€å¤§å€¼
# axis=2 = æ²¿è‘—ç¬¬2å€‹è»¸å–æœ€å¤§å€¼
gray_max = np.max(astronaut_rgb, axis=2).astype(np.uint8)

print("ä¸‰ç¨®è½‰æ›æ–¹æ³•:")
print("1. æ¨™æº–åŠ æ¬Š: 0.299Ã—R + 0.587Ã—G + 0.114Ã—B")
print("2. ç°¡å–®å¹³å‡: (R + G + B) Ã· 3")
print("3. å–æœ€å¤§å€¼: max(R, G, B)")
print()

# é¡¯ç¤ºæ¯”è¼ƒçµæœ
plt.figure(figsize=(20, 5))  # å¾ˆå¯¬çš„åœ–

# åŸå§‹RGBåœ–åƒ
plt.subplot(1, 4, 1)
plt.imshow(astronaut_rgb)
plt.title('åŸå§‹ RGB åœ–åƒ')
plt.axis('off')

# æ¨™æº–åŠ æ¬Šè½‰æ›
plt.subplot(1, 4, 2)
plt.imshow(gray_standard, cmap='gray')  # cmap='gray' ç”¨æ–¼é¡¯ç¤ºç°åº¦åœ–
plt.title('æ¨™æº–åŠ æ¬Šè½‰æ›\n(è€ƒæ…®äººçœ¼æ•æ„Ÿåº¦)')
plt.axis('off')

# ç°¡å–®å¹³å‡è½‰æ›
plt.subplot(1, 4, 3)
plt.imshow(gray_average, cmap='gray')
plt.title('ç°¡å–®å¹³å‡è½‰æ›\n(R+G+B)/3')
plt.axis('off')

# æœ€å¤§å€¼è½‰æ›
plt.subplot(1, 4, 4)
plt.imshow(gray_max, cmap='gray')
plt.title('æœ€å¤§å€¼è½‰æ›\nmax(R,G,B)')
plt.axis('off')

plt.tight_layout()
plt.show()

print("ğŸ” è§€å¯Ÿå·®ç•°:")
print("- æ¨™æº–è½‰æ›: æœ€è‡ªç„¶ï¼Œä¿æŒè¦–è¦ºå°æ¯”åº¦")
print("- ç°¡å–®å¹³å‡: ç¨å¾®åäº®ï¼Œä¸å¤ªè‡ªç„¶")
print("- æœ€å¤§å€¼: éäº®ï¼Œå¤±å»ç´°ç¯€")
print("- æ¨è–¦ä½¿ç”¨æ¨™æº–åŠ æ¬Šæ–¹æ³•")
```

## 5ï¸âƒ£ æ•¸çµ„åˆ‡ç‰‡èˆ‡åœ–åƒå€åŸŸæ“ä½œ

### ğŸ¯ é—œéµæ¦‚å¿µè§£é‡‹
- **åˆ‡ç‰‡èªæ³• [start:end]** = å¾startåˆ°endï¼ˆä¸åŒ…å«endï¼‰
- **[:10]** = å¾é–‹å§‹åˆ°ç¬¬10å€‹ï¼ˆä¸å«ï¼‰
- **[::2]** = æ¯éš”ä¸€å€‹å–æ¨£ï¼ˆæ­¥é•·ç‚º2ï¼‰
- **[è¡Œèµ·:è¡Œçµ‚, åˆ—èµ·:åˆ—çµ‚]** = æå–çŸ©å½¢å€åŸŸ

```python
# ä½¿ç”¨æ¨™æº–ç°åº¦åœ–åƒé€²è¡Œåˆ‡ç‰‡ç·´ç¿’
gray_img = np.dot(astronaut_rgb[...,:3], [0.299, 0.587, 0.114]).astype(np.uint8)

print("=== æ•¸çµ„åˆ‡ç‰‡èªæ³•è©³ç´°æ•™å­¸ ===")
print(f"å®Œæ•´åœ–åƒå½¢ç‹€: {gray_img.shape}")  # (512, 512)
print()

print("åˆ‡ç‰‡èªæ³•è¦å‰‡:")
print("array[è¡Œ, åˆ—] - ç”¨æ–¼2Dæ•¸çµ„")
print("start:end - å¾startåˆ°endï¼ˆä¸åŒ…å«endï¼‰")
print(": - è¡¨ç¤ºå…¨éƒ¨")
print(":10 - å¾é–‹å§‹åˆ°ç¬¬10å€‹")
print("::2 - æ¯éš”ä¸€å€‹å–æ¨£")
print()

# æå–ä¸åŒå¤§å°çš„å€åŸŸ
print("=== æå–ä¸åŒå€åŸŸ ===")

# [:3, :3] = å‰3è¡Œï¼Œå‰3åˆ—
patch_3x3 = gray_img[:3, :3]
print(f"å·¦ä¸Šè§’3x3å€åŸŸ: gray_img[:3, :3] â†’ å½¢ç‹€ {patch_3x3.shape}")

# [:10, :10] = å‰10è¡Œï¼Œå‰10åˆ—
patch_10x10 = gray_img[:10, :10]
print(f"å·¦ä¸Šè§’10x10å€åŸŸ: gray_img[:10, :10] â†’ å½¢ç‹€ {patch_10x10.shape}")

# [231:281, 231:281] = ç¬¬231-280è¡Œï¼Œç¬¬231-280åˆ—ï¼ˆä¸­å¿ƒå€åŸŸï¼‰
center_50x50 = gray_img[231:281, 231:281]
print(f"ä¸­å¿ƒ50x50å€åŸŸ: gray_img[231:281, 231:281] â†’ å½¢ç‹€ {center_50x50.shape}")

print()
print("=== 3x3å€åŸŸçš„å¯¦éš›åƒç´ å€¼ ===")
print("æ¯å€‹æ•¸å­—ä»£è¡¨ä¸€å€‹åƒç´ çš„äº®åº¦å€¼:")
print(patch_3x3)
```

### ğŸ’¡ é€²éšåˆ‡ç‰‡ï¼šé™é‡‡æ¨£ï¼ˆDownsamplingï¼‰

```python
# === æ­¥é•·åˆ‡ç‰‡ - é™é‡‡æ¨£æŠ€è¡“ ===

print("=== æ­¥é•·åˆ‡ç‰‡èªæ³• ===")
print("array[start:end:step] - stepæ˜¯æ­¥é•·")
print("::2 - æ¯éš”1å€‹å–æ¨£ï¼ˆæ­¥é•·2ï¼‰")
print("::4 - æ¯éš”3å€‹å–æ¨£ï¼ˆæ­¥é•·4ï¼‰")
print()

# [::2, ::2] = è¡Œå’Œåˆ—éƒ½æ¯éš”ä¸€å€‹å–æ¨£
downsampled_2x = gray_img[::2, ::2]

# [::4, ::4] = è¡Œå’Œåˆ—éƒ½æ¯éš”ä¸‰å€‹å–æ¨£
downsampled_4x = gray_img[::4, ::4]

print("=== é™é‡‡æ¨£çµæœ ===")
print(f"åŸå§‹åœ–åƒ: {gray_img.shape} = {gray_img.size} åƒç´ ")
print(f"2å€é™é‡‡æ¨£: {downsampled_2x.shape} = {downsampled_2x.size} åƒç´ ")
print(f"4å€é™é‡‡æ¨£: {downsampled_4x.shape} = {downsampled_4x.size} åƒç´ ")
print()

# è¨ˆç®—æ•¸æ“šå£“ç¸®ç‡
compression_2x = downsampled_2x.size / gray_img.size
compression_4x = downsampled_4x.size / gray_img.size
print(f"2å€é™é‡‡æ¨£ä¿ç•™ {compression_2x:.1%} çš„åƒç´ ")
print(f"4å€é™é‡‡æ¨£ä¿ç•™ {compression_4x:.1%} çš„åƒç´ ")

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(gray_img, cmap='gray')
plt.title(f'åŸå§‹åœ–åƒ\n{gray_img.shape[0]}Ã—{gray_img.shape[1]}')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(downsampled_2x, cmap='gray')
plt.title(f'2å€é™é‡‡æ¨£\n{downsampled_2x.shape[0]}Ã—{downsampled_2x.shape[1]}')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(downsampled_4x, cmap='gray')
plt.title(f'4å€é™é‡‡æ¨£\n{downsampled_4x.shape[0]}Ã—{downsampled_4x.shape[1]}')
plt.axis('off')

plt.tight_layout()
plt.show()

print("ğŸ“Š é™é‡‡æ¨£çš„ç”¨é€”:")
print("- æ¸›å°‘åœ–åƒå¤§å°ï¼Œç¯€çœå­˜å„²ç©ºé–“")
print("- åŠ å¿«è™•ç†é€Ÿåº¦")
print("- å‰µå»ºåœ–åƒé‡‘å­—å¡”ï¼ˆå¤šå°ºåº¦åˆ†æï¼‰")
print("- ä½†æœƒæå¤±åœ–åƒç´°ç¯€")
```
